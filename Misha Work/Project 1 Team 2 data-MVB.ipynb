```python
import pandas as pd
```

```python
df = pd.read_csv("MaryJane_clean_data.csv")
df.head()
```

```python
df.Effects.str.split(",",expand=True)
```

```python
effects = df.Effects.str.split(",",expand=True)
effects.columns=["effect 1", "effect 2", "effect 3", "effect 4", "effect 5"]
effects.head()
```

```python
flavors = df.Flavor.str.split(",",expand=True)
flavors.columns=["flavor 1", "flavor 2", "flavor 3", "flavor 4"]
flavors.head()
```

```python
flavors.info()
```

```python
pd.concat([df, effects, flavors], axis=1)
```

```python
df2=pd.concat([df, effects, flavors], axis=1)
df2.head()
```

```python
df2["num_effects"]=df2.Effects.str.count(",")+1
df2["num_flavors"]=df2.Flavor.str.count(",")+1
df2.info()
```
